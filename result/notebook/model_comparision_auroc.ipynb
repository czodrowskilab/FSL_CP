```python
# Avg score of each method for each assay
avg_proto_cp = np.mean([auroc_proto_cp_8,auroc_proto_cp_16, auroc_proto_cp_32, auroc_proto_cp_64, auroc_proto_cp_96], axis=0)
avg_proto_img = np.mean([auroc_proto_img_8, auroc_proto_img_16, auroc_proto_img_32, auroc_proto_img_64, auroc_proto_img_96], axis=0)
avg_multitask_cp = np.mean([auroc_multitask_cp_8, auroc_multitask_cp_16, auroc_multitask_cp_32, auroc_multitask_cp_64, auroc_multitask_cp_96], axis=0)
avg_maml_cp = np.mean([auroc_maml_cp_8, auroc_maml_cp_16, auroc_maml_cp_32, auroc_maml_cp_64, auroc_maml_cp_96], axis=0)
avg_xgboost_cp = np.mean([auroc_xgboost_cp_8, auroc_xgboost_cp_16, auroc_xgboost_cp_32, auroc_xgboost_cp_64, auroc_xgboost_cp_96], axis=0)
avg_maml_img = np.mean([auroc_maml_img_8, auroc_maml_img_16, auroc_maml_img_32, auroc_maml_img_64, auroc_maml_img_96], axis=0)
labels = list(df_proto_auroc_cp['assay_chembl_id'])

with plt.style.context(['seaborn-v0_8-deep', 'seaborn']):
    x = np.arange(len(labels))  # the label locations
    width = 0.15  # the width of the bars

    fig, ax = plt.subplots()
    fig.set_size_inches(15, 6)
    #rects1 = ax.bar(x - width, avg_maml_cp, width, label='maml')
    rects2 = ax.bar(x-width, avg_proto_cp, width, label='protonet_cp')
    rects3 = ax.bar(x, avg_proto_img, width, label='protonet_img')
    #rects4 = ax.bar(x + width, avg_xgboost_cp, width, label='xgboost_cp')
    rects5 = ax.bar(x + 2*width, avg_multitask_cp, width, label='multitask_cp')
    rects6 = ax.bar(x + width, avg_maml_img, width, label='maml_img')
    
    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_ylabel('AUC')
    ax.set_title('Mean AUC of unseen assay (Top 4 models)')
    ax.set_xticks(x, labels)
    ax.legend(loc='upper left')

    #ax.bar_label(rects1, padding=3)
    #ax.bar_label(rects2, padding=3)

    fig.tight_layout()

    plt.tick_params(axis='x', labelrotation=90)

    plt.ylim([0.45, 0.76])
    plt.savefig("../plots/Model_comparison_barplot1.png", bbox_inches='tight')
    plt.show()
```

```python
avg_protonet_cp = [np.mean(auroc_proto_cp_8), np.mean(auroc_proto_cp_16), np.mean(auroc_proto_cp_32), np.mean(auroc_proto_cp_64), np.mean(auroc_proto_cp_96)]
avg_protonet_img = [np.mean(auroc_proto_img_8), np.mean(auroc_proto_img_16), np.mean(auroc_proto_img_32), np.mean(auroc_proto_img_64), np.mean(auroc_proto_img_96)]
avg_multitask_cp = [np.mean(auroc_multitask_cp_8), np.mean(auroc_multitask_cp_16), np.mean(auroc_multitask_cp_32), np.mean(auroc_multitask_cp_64), np.mean(auroc_multitask_cp_96)]
avg_maml_cp = [np.mean(auroc_maml_cp_8), np.mean(auroc_maml_cp_16), np.mean(auroc_maml_cp_32), np.mean(auroc_maml_cp_64), np.mean(auroc_maml_cp_96)]
avg_xgboost_cp = [np.mean(auroc_xgboost_cp_8), np.mean(auroc_xgboost_cp_16), np.mean(auroc_xgboost_cp_32), np.mean(auroc_xgboost_cp_64), np.mean(auroc_xgboost_cp_96)]
avg_maml_img = [np.mean(auroc_maml_img_8), np.mean(auroc_maml_img_16), np.mean(auroc_maml_img_32), np.mean(auroc_maml_img_64), np.mean(auroc_maml_img_96)]

labels = ['8', '16', '32', '64', '96']

with plt.style.context(['seaborn-v0_8-deep', 'seaborn']):
    plt.errorbar(labels, avg_multitask_cp, yerr=np.std(avg_multitask_cp), color='orange', label='multitask_cp', marker='o', linewidth=1)
    plt.errorbar(labels, avg_protonet_cp, yerr=np.std(avg_protonet_cp), color='red', label='protonet_cp', marker='o', linewidth=1)
    plt.errorbar(labels, avg_protonet_img, yerr=np.std(avg_protonet_img), color='red', linestyle='dashed', label='protonet_img', marker='o', linewidth=1)
    plt.errorbar(labels, avg_maml_cp, yerr=np.std(avg_maml_cp), color='blue', label='maml_cp', marker='o', linewidth=1)
    plt.errorbar(labels, avg_xgboost_cp, yerr=np.std(avg_xgboost_cp), color='green', label='xgboost_cp', marker='o', linewidth=1)
    plt.errorbar(labels, avg_maml_img, yerr=np.std(avg_maml_img), color='blue', label='maml_img', marker='o', linestyle='dashed', linewidth=1)

    # Add some text for labels, title and custom x-axis tick labels, etc.
    plt.ylabel('AUC')
    plt.xlabel('Support set size')
    plt.suptitle('Mean AUC on unseen assays as support set size increases')
    #plt.xticks([])
    plt.legend(loc='upper left')

    #ax.bar_label(rects1, padding=3)
    #ax.bar_label(rects2, padding=3)

    fig.tight_layout()

    plt.tick_params(axis='x', labelrotation=90)

    plt.ylim([0.475, 0.64])
    plt.savefig("../plots/Model_comparison_lineplot1.png")
    plt.show()
```
