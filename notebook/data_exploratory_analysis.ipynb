```python
import pandas as pd
import matplotlib.pyplot as plt
import os 
import sys

FSL_CP_PATH = os.path.join(os.environ['HOME'], 'FSL_CP')
os.chdir(FSL_CP_PATH)
```

```python
# Load data
df_label = pd.read_csv('data/output/FINAL_LABEL_DF.csv', dtype={"ASSAY":"string"})
df_map = pd.read_csv('data/output/assay_target_map.csv')
```

```python
# Dataset statistics
print("In this dataset,...")
print(f"...there are {len(set(df_label['ASSAY']))} assays.")
print(f"...there are {len(set(df_label['CPD_SMILES']))} compounds.")
print(f"...there are {len(set(df_map['target_chembl_id']))} different protein targets.")
```

```python
# duplicates by smiles
with plt.style.context(['seaborn-v0_8-deep', 'seaborn']):
    plt.hist(df_label.drop_duplicates(subset='SAMPLE_KEY')['CPD_SMILES'].value_counts(), bins='auto')
    plt.ylabel('Count')
    plt.xlabel('Number of duplicates')
    plt.suptitle('Distribution of number of compound duplicates')
```
